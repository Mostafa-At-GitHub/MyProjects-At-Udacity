```python
# this cell for demonstrating using melt() function to make treatment and start_end columns

ef = pd.melt(treatments_clean, id_vars=["given_name","surname","hba1c_start","hba1c_end","hba1c_change"], 
                 var_name="treatment", value_name="start_end")
ef=ef.query('start_end !="-"')
ef['start_dose'], ef['end_dose'] = ef['start_end'].str.split(' - ', 1).str   #more concise way to create start and end columns
ef = ef.drop('start_end', axis=1)
ef.reset_index(drop=True,inplace=True)
ef.sample(5)
```

```python
# Your cleaning code here                                           #my solution using for loop without melt
# creating treatment,start_dose, end_dose columns
treatment_series=[]
start_dose_series=[]
end_dose_series= []

for index,value in treatments_clean.auralin.items():
    if value != '-':
        treatment_series.append('auralin')
        start_dose_series.append(treatments_clean.auralin[index].split(' - ')[0].strip('u'))
        end_dose_series.append(treatments_clean.auralin[index].split(' - ')[1].strip('u'))
    else:
        treatment_series.append('novodra')
        start_dose_series.append(treatments_clean.novodra[index].split(' - ')[0].strip('u'))
        end_dose_series.append(treatments_clean.novodra[index].split(' - ')[1].strip('u'))
        
treatments_clean['treatment']=treatment_series
treatments_clean['start_dose']=start_dose_series
treatments_clean['end_dose']=end_dose_series 
```

#### Full state names sometimes, abbreviations other times

##### Define
*Your definition here. Hint: [tutorial](https://chrisalbon.com/python/pandas_apply_operations_to_dataframes.html) for method used in solution.*

- Change names in state column to abbreviations as follows:
  - California to CA
  - New York to NY
  - Illinois to IL
  - Florida to FL
  - Nebraska to NE


##### Code

```python
patients_clean.state.value_counts()
```

```python
# Your cleaning code here              #this is my solution

#patients_clean_copy=patients_clean.copy()

s_names={"California":'CA',"New York":'NY',"Illinois":'IL',"Florida":'FL',"Nebraska":'NE'}

for index,value in patients_clean.state.items():
    if value in s_names:
        patients_clean.state[index] = s_names[value]
```

```python
#official solution
# Mapping from full state name to abbreviation
state_abbrev = {'California': 'CA',
                'New York': 'NY',
                'Illinois': 'IL',
                'Florida': 'FL',
                'Nebraska': 'NE'}

# Function to apply
def abbreviate_state(patient):
    if patient['state'] in state_abbrev.keys():
        abbrev = state_abbrev[patient['state']]
        return abbrev
    else:
        return patient['state']
    
patients_clean['state'] = patients_clean.apply(abbreviate_state, axis=1)
```
