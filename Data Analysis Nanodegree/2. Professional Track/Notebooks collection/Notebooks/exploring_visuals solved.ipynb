# Exploring with Visuals
Use `clean_08.csv` and `clean_18.csv`. You should've created these data files in the previous section: *Fixing Data Types Pt 3*.

```python
# load datasets
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#import seaborn as sns

df_08=pd.read_csv('clean_08.csv')
df_18=pd.read_csv('clean_18.csv')
```

## 1st question. Distributions of greenhouse gas score in 2008 & 2018

### Visualizing by `matplotlib.pyplot.plot.hist()`

#### 1.1 Combined histograms by `matplotlib.pyplot.plot.hist()`

```python
plt.hist(df_08.greenhouse_gas_score)
plt.hist(df_18.greenhouse_gas_score,alpha=.8)
plt.title('Distributions of greenhouse gas score in 2008 & 2018',size=15)
plt.xlabel('Score',size=15)
plt.ylabel('Frequency',size=15)
plt.legend(['2008','2018']);
```

#### 1.2 Individual histograms  by `matplotlib.pyplot.plot.hist()`

```python
plt.hist(df_08.greenhouse_gas_score)
plt.title('Distribution of greenhouse gas score in 2008',size=15)
plt.xlabel('Score',size=15)
plt.ylabel('Frequency',size=15);
```

```python
plt.hist(df_18.greenhouse_gas_score)
plt.title('Distribution of greenhouse gas score in 2018',size=15)
plt.xlabel('Score',size=15)
plt.ylabel('Frequency',size=15);
```

### Visualizing by `pandas.DataFrame.hist()`

#### 1.1 Combined histograms by `pandas.DataFrame.hist()`

```python
df_08.greenhouse_gas_score.hist(),df_18.greenhouse_gas_score.hist(alpha=.7);
```

#### 1.2 Individual histograms by `pandas.DataFrame.hist()`

```python
df_08.greenhouse_gas_score.hist();    #2008
```

```python
df_18.greenhouse_gas_score.hist();    #2018
```

## 2nd  question. Distributions of combined mpg in 2008 & 2018

```python
plt.hist(df_08.cmb_mpg,)
plt.hist(df_18.cmb_mpg,alpha=.72)
plt.title('Distributions of combined mpg in 2008 & 2018',size=15)
plt.xlabel('Score',size=15)
plt.ylabel('Frequency',size=15)
plt.legend(['2008','2018']);
```

## 3rd question. Correlation between Displacement and Combined MPG in 2008

```python
plt.scatter(df_08.displ,df_08.cmb_mpg);
plt.title('Correlation between Displacement and Combined MPG',size=15)
plt.xlabel('Displacement',size=15)
plt.ylabel('Combined MPG',size=15);
```

## 4th question. Correlation between Greenhouse gas score and Combined MPG in 2008

```python
plt.scatter(df_08.greenhouse_gas_score,df_08.cmb_mpg);
plt.title('Correlation between Greenhouse gas score and Combined MPG',size=15)
plt.xlabel('Greenhouse gas score',size=15)
plt.ylabel('Combined MPG',size=15);
```
